```python
#tickers list = XLB, XLE, XLF, XLI, XLK, XLP, XLU, XLV, XLY
tickers_list = ['XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV', 'XLY']

yfobj= yf.Tickers(tickers_list)
#start from 2015-01-01
yf_hist = yfobj.history(start="2015-01-01", end="2024-01-01")
yf_hist

#keep only the Close
yf_hist = yf_hist[['Close']]

#ffill
yf_hist = yf_hist.fillna(method='ffill')
#drop Close level
yf_hist.columns = yf_hist.columns.droplevel(0)

#resample to weekly 
yf_hist = yf_hist.resample('W').last()
yf_hist
```

```python
#plot the data
import matplotlib.pyplot as plt
plt.plot(yf_hist)

plt.legend(yf_hist.columns)
plt.show()
```

```python
#return calc
rets = yf_hist.pct_change().fillna(0)
rets

#visualize the data of rets
import matplotlib.pyplot as plt
plt.plot(rets)
plt.show()
```

```python
#rolling volatility (1ë…„ë‹¨ìœ„ ë¡¤ë§ë³€ë™ì„±ê³„ì‚° ë° ì—°ìœ¨í™”)
roll_vol = rets.rolling(window=52).std() * np.sqrt(52)
roll_vol
#visualize the data of roll_vol
import matplotlib.pyplot as plt
plt.plot(roll_vol)
plt.show()
```

```python
#total period correlation coff
tot_corr = rets.corr()
tot_corr
```

```python
import seaborn as sns
sns.heatmap(tot_corr, annot=True)
plt.show()
```

```python
#covariance annualized
cov = rets.cov() * 52
cov
```

```python
#using var() method to calculate the variance of each asset
var = rets.var() * 52
var
```

```python
#diag() method to extract the diagonal values of the covariance matrix
cov_diag = cov.values.diagonal()
cov_diag
```

```python
#ticker = SPY, TLT
tickers_list = ['SPY', 'TLT']
#start from 2002-01-01
yfobj= yf.Tickers(tickers_list)
yf_hist = yfobj.history(start="2002-01-01", end="2024-01-01")
#Keep only the Close for SPY and TLT
yf_hist = yf_hist[['Close']]
#Drop close level
yf_hist.columns = yf_hist.columns.droplevel(0)
#plot the data
import matplotlib.pyplot as plt
plt.plot(yf_hist)
plt.legend(yf_hist.columns)
plt.show()
```

```python
#return calculation
rets = yf_hist.pct_change().fillna(0)
rets
#visualize the data of rets
import matplotlib.pyplot as plt
plt.plot(rets)
plt.show()
```

```python
#individual stock average return and volatiltiy, sharpe ratio
stock_mu = rets['SPY'].mean() * 252
stock_vol = rets['SPY'].std() * np.sqrt(252)
stock_sharpe = stock_mu / stock_vol
print(stock_sharpe)
#individual bond average return and volatiltiy, sharpe ratio
bond_mu = rets['TLT'].mean() * 252
bond_vol = rets['TLT'].std() * np.sqrt(252)
bond_sharpe = bond_mu / bond_vol
print(bond_sharpe)
```

```python
#60-40 portfolio

weight = 0.6
stock_weight = weight
bond_weight = 1 - weight
#portfolio return
portfolio_mu = stock_weight * rets['SPY'] + bond_weight * rets['TLT']
portfolio_mu.dropna(inplace=True)
portfolio_mu
#cumulative return
portfolio_cum = (1 + portfolio_mu).cumprod()
#viosualize the data of portfolio_cum, SPY, TLT
import matplotlib.pyplot as plt
plt.plot(portfolio_cum)
plt.plot((1+rets['SPY']).cumprod())
plt.plot((1+rets['TLT']).cumprod())
plt.legend(['Portfolio', 'SPY', 'TLT'])
plt.show()
```
